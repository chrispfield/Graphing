I then constructed a basic line graph. This graph exhibits the total annual imports (blue) and exports (yellow) of trout products between 1989 and 2018. Note that there is a significant difference between the two, with the import volumes steadily increasing and the export volumes remaining fairly steady. This indicates a large trade deficit. There are several possible reasons for the increase in import volumes while export volumes stagnated:

1. Strong demand in the USA results in better prices locally than internationally, and so although local production is increasing at the same rate as import rate, producers preferentially sell locally.

2. Local and international prices remain equal, but production has not increased significantly, thus imports have increased to meet demand.

Whatever the reason, it appears that there is a strong demand for international trout products and so international producers should explore it as a potential market. It would be interesting to include the amount of trout produced for local consumption, as this would give a more comprehensive picture and help elucidate why there has been an increase in the trade deficit for this category of products.

I then constructed a series of subplots to see how the export volumes have changed over the previous 3 decades. Both salmon and shrimp volumes have decreased slightly over the last 15 years. This may be due to Norway and Scotland capturing the international salmon products market and Southeast Asian countries capturing the international shrimp products market. In contrast, shellfish products saw significant growth over the period 15 to 5 years ago. The market has stagnated since then, but this growth exhibits the sudden consumer interest in shellfish as a food product during that period. Other countries have since started ramping up shellfish production and it may be that the USA will need to support this local industry if they want to continue to see growth in this sector. Trout product exports have very recently seen a marked growth, it would be interesting to see if this trajectory continues. 

```python
sns.set(rc = {'figure.figsize':(20,11)})
sns.set_style('whitegrid')

fig, axes = plt.subplots(2, 2)

fig.set_facecolor((0.05, 0.05, 0.12))
axes[0,0].set_facecolor((0.05, 0.05, 0.12))
axes[0,1].set_facecolor((0.05, 0.05, 0.12))
axes[1,0].set_facecolor((0.05, 0.05, 0.12))
axes[1,1].set_facecolor((0.05, 0.05, 0.12))

fig.subplots_adjust(hspace=0.3, wspace=0.2)


sns.lineplot(ax=axes[0,0],x=df_grouped.index, y=df_grouped['Trout, fresh and frozen'])
axes[0,0].set_ylabel('Tonnes', color='white', fontsize=15, fontname='Arial')
axes[0,0].set_xlabel('Year', color='white', fontsize=15, fontname='Arial')
axes[0,0].set_ylim(bottom=0, top=3900)
axes[0,0].set_xlim(left=1989, right=2019)
axes[0,0].set_title('Trout, fresh and frozen', color='white',fontsize=17)

axes[0,0].tick_params(colors='white')


sns.lineplot(ax=axes[0,1],x=df_grouped.index, y=df_grouped['Salmon'])
axes[0,1].set_ylabel('Tonnes', color='white', fontsize=15, fontname='Arial')
axes[0,1].set_xlabel('Year', color='white', fontsize=15, fontname='Arial')
axes[0,1].set_ylim(bottom=0)
axes[0,1].set_xlim(left=1989, right=2019)
axes[0,1].set_title('Salmon', color='white',fontsize=17)

axes[0,1].tick_params(colors='white')

sns.lineplot(ax=axes[1,0],x=df_grouped.index, y=df_grouped['Shrimp'])
axes[1,0].set_ylabel('Tonnes', color='white', fontsize=15, fontname='Arial')
axes[1,0].set_xlabel('Year', color='white', fontsize=15, fontname='Arial')
axes[1,0].set_ylim(bottom=0)
axes[1,0].set_xlim(left=1989, right=2019)
axes[1,0].set_title('Shrimp', color='white',fontsize=17)

axes[1,0].tick_params(colors='white')

sns.lineplot(ax=axes[1,1],x=df_grouped.index, y=df_grouped['Shellfish'])
axes[1,1].set_ylabel('Tonnes', color='white', fontsize=15, fontname='Arial')
axes[1,1].set_xlabel('Year', color='white', fontsize=15, fontname='Arial')
axes[1,1].set_ylim(bottom=0)
axes[1,1].set_xlim(left=1989, right=2019)
axes[1,1].set_title('\nShellfish', color='white',fontsize=17)

axes[1,1].tick_params(colors='white')

fig.suptitle('Total annual export of various aquaculture food products by weight', fontsize=22, fontweight='bold', fontname='Helvetica', color='white')
                                                                                             
sns.despine()                                                                                                                       
```

I then constructed a stacked bar chart of the above information.
